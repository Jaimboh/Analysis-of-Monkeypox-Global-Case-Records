```python
# Extract data for Kenya
uganda_data = df[df['location'] == 'Uganda']
```

```python
uganda_dates = uganda_data['date'].tolist()
```

```python
uganda_new_cases = uganda_data['new_cases'].tolist()
```

```python
line = (
    Line()
    .add_xaxis(uganda_dates)  # Add x-axis (dates for Uganda)
    .add_yaxis("Uganda New Cases", uganda_new_cases, is_smooth=True,
               label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (new cases in Uganda)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Uganda Daily New Cases"),
        xaxis_opts=opts.AxisOpts(name="Date"),
        yaxis_opts=opts.AxisOpts(name="New Cases"),
        toolbox_opts=opts.ToolboxOpts()
    )
)

# line.render("uganda_new_cases.html")
line.render_notebook()
```

```python
uganda_total_cases = uganda_data['total_cases'].tolist()

line2 = (
    Line()
    .add_xaxis(uganda_dates)  # Add x-axis (dates for Uganda)
    .add_yaxis("Uganda Total Cases", uganda_total_cases, is_smooth=True,
               label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (total cases in Uganda)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Uganda Cumulative Cases"),
        xaxis_opts=opts.AxisOpts(name="Date"),
        yaxis_opts=opts.AxisOpts(name="Total Cases"),
        toolbox_opts=opts.ToolboxOpts()
    )
)

# line2.render("Uganda_total_cases.html")
line2.render_notebook()
```

```python
uganda_new_cases_per_million = uganda_data['new_cases_per_million'].tolist()

line_per_million = (
    Line()
    .add_xaxis(uganda_dates)  # Add x-axis (dates for Uganda)
    .add_yaxis("Uganda New Cases per Million", uganda_new_cases_per_million, is_smooth=True,
               label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (new cases per million in Uganda)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Uganda New Cases per Million"),
        xaxis_opts=opts.AxisOpts(name="Date"),
        yaxis_opts=opts.AxisOpts(name="New Cases per Million"),
        toolbox_opts=opts.ToolboxOpts()
    )
)

# line_per_million.render("uganda_new_cases_per_million.html")
line_per_million.render_notebook()
```

```python
world_data = df[df['location'] == 'World']

# Extract dates, total cases, and total deaths
dates = world_data['date'].tolist()
total_cases = world_data['total_cases'].tolist()
total_deaths = world_data['total_deaths'].tolist()

line_total = (
    Line()
    .add_xaxis(dates)  # Add x-axis (global dates)
    .add_yaxis("Total Cases", total_cases, is_smooth=True, label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (total global cases)
    .add_yaxis("Total Deaths", total_deaths, is_smooth=True, label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (total global deaths)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Global Total Cases and Deaths Over Time"),
        xaxis_opts=opts.AxisOpts(name="Date"),
        yaxis_opts=opts.AxisOpts(name="Count"),
        toolbox_opts=opts.ToolboxOpts(),
    )
)

# line_total.render("global_total_cases_and_deaths.html")
line_total.render_notebook()
```

```python
# Extract total cases and total deaths per million
total_cases_per_million = world_data['total_cases_per_million'].tolist()
total_deaths_per_million = world_data['total_deaths_per_million'].tolist()

line_per_million = (
    Line()
    .add_xaxis(dates)  # Add x-axis (global dates)
    .add_yaxis("Total Cases per Million", total_cases_per_million, is_smooth=True,
               label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (total cases per million globally)
    .add_yaxis("Total Deaths per Million", total_deaths_per_million, is_smooth=True,
               label_opts=opts.LabelOpts(is_show=True))  # Add y-axis (total deaths per million globally)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Global Total Cases and Deaths per Million Over Time"),
        xaxis_opts=opts.AxisOpts(name="Date"),
        yaxis_opts=opts.AxisOpts(name="Count per Million"),
        toolbox_opts=opts.ToolboxOpts(),
    )
)

# Render the chart
# line_per_million.render("global_total_cases_and_deaths_per_million.html")
line_per_million.render_notebook()


```
